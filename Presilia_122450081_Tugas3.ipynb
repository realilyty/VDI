**Proses Visualisasi Data**

```python
#import library yang akan digunakan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dan baca data
df=pd.read_csv('vdi.csv')
df
```

```python
# Melakukan pemebersihan disini menghapus baris dengan indeks 0 dan 1 karna tidak dipakai
df_clean = df.drop([0, 1])

# Mengatur ulang indeks setelah penghapusan
df_clean = df_clean.reset_index(drop=True)

# Menampilkan DataFrame setelah baris 0 dan 1 dihapus
df_clean
```

```python
# Mengganti nama kolom menjadi "Indikator Indeks Demokrasi Indonesia", "2022", dan "2021"
df_clean.columns = ['Indikator Indeks Demokrasi Indonesia', '2022', '2021']

# Menampilkan DataFrame dengan nama kolom yang sudah diubah
df_clean
```

```python
#Mengubah Indikator IDI menjadi angka agar mudah divisualisasikan
df_clean['Indikator Indeks Demokrasi Indonesia'] = [str(i+1) for i in range(len(df_clean))]
df_clean
```

```python
# Create a bar plot of values for each year, grouped by indicator
sns.barplot(x="Indikator Indeks Demokrasi Indonesia", y="2021", data=df_clean, label="2021", color='navy')
plt.title("Bar Plot berdasarkan Value Tiap Indikator")
plt.legend()
plt.show()
```

```python
sns.barplot(x="Indikator Indeks Demokrasi Indonesia", y="2022", data=df_clean, label="2022", color='navy')
plt.title("Bar Plot berdasarkan Value Tiap Indikator")
plt.legend()
plt.show()
```
